```python
punctuation = '.,:;-!?"'
for mark in punctuation:
    data = data.replace(mark, '')
    
```

```python
words = data.lower().strip().split()
words
```

```python
count = len(words)
print(count)
word_counts = dict()
```

```python
x = [1,1,2,2,3,3,3,4,5]
```

```python
set(x)
```

```python
for word in words:
    if word not in word_counts:
        word_counts[word] = 1
    else:
        word_counts[word] += 1
```

```python
unique_words = len(word_counts)
print(unique_words)
```

```python
proportion = unique_words / count
print(proportion)
```

```python
from operator import itemgetter
sorted_counts = sorted(word_counts.items(), key=itemgetter(1), reverse = True)
```

```python
print(f'{"Word Counts":^20}')
print()
print(f"{'Word':<15}{'Count':>5}")
print('-' * 20)
for word, value in sorted_counts[:15]:
    print(f"{word:15}{value:>5}")
print(f"{'Total words:':23} {count:>5}")
print(f"Unique words: {unique_words}")
print(f"Proportion unique words: {proportion:>.2f}")
```
