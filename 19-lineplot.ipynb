# Basic Line Plot with Python Matplotlib

### 1. Set up environment - here we will import stock price data using pandas_datareader


```python
import pandas as pd
import pandas_datareader as pdr
import matplotlib.pyplot as plt
```

### 2. Get some data

```python
stocks = ['GOOG', 'AMZN']
data = pdr.get_data_yahoo(stocks, start = '2019-01-01')['Close']
data.head()
```

### 3. Plot Data 

```python
from matplotlib import rcParams
rcParams['figure.figsize'] = 12,6
plt.plot(data.AMZN, label = 'AMZN')
plt.plot(data.GOOG, label = 'GOOG')
plt.grid(True, color='k', linestyle=':')
plt.title("Amazon & Google Prices")
plt.xlabel("Date")
plt.ylim(500, 3500)
plt.xlim('2019-01-01', '2021-04-30')
plt.xticks(['2019-03','2019-06', '2019-12', '2021-04'])
plt.tick_params(axis = 'both', labelsize = 10)
plt.yticks([500, 1000,1500,2000,2500,3000, 3500])
plt.style.use('fivethirtyeight')
plt.legend(loc=2);
```
